### What is data interpolation?
Data interpolation is a technique used to estimate values within a dataset based on the known values. It's particularly useful when you have missing data points or want to fill in gaps between existing data points. Interpolation assumes that the data between two known points follows a certain pattern, allowing you to make educated guesses about the values at those intermediate points.

There are different methods of interpolation, including:
1. **Linear Interpolation:**
   This method assumes a straight-line relationship between data points. It calculates the values between two known points based on the slope of the line connecting them.

2. **Polynomial Interpolation:**
   Polynomial interpolation fits a polynomial function through the known data points. Higher-degree polynomials can capture more complex relationships, but they can also lead to overfitting.

3. **Spline Interpolation:**
   Spline interpolation uses piecewise polynomial functions to create a smooth curve that passes through the known data points. It avoids the issue of oscillation that can occur with higher-degree polynomial interpolation.

4. **Nearest-Neighbor Interpolation:**
   Nearest-neighbor interpolation assigns the value of the nearest known data point to the missing or interpolated point. It's simple but might not capture complex relationships.

5. **Inverse Distance Weighting:**
   This method calculates an interpolated value by considering the distances and values of nearby known data points. Closer points have more influence on the interpolated value.

6. **Kriging (Geostatistical Interpolation):**
   Kriging is used in spatial interpolation, particularly in geostatistics. It models spatial dependence between data points to create more accurate interpolations.

Each method has its own strengths and limitations, and the choice of interpolation method depends on the nature of your data and the context in which you're working.

### 1. Linear Interpolation

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
# i'll create two array's of size 5
x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])
```

```python
x, y
```

```python
plt.scatter(x, y)
```

```python
# now i'll interpolate my data using linear interpolation
x_new = np.linspace(1, 5, 10) #create new x values
y_interp = np.interp(x_new, x, y) #create new y values
```

```python
plt.scatter(x_new, y_interp)
```

### 2. Cubic Interpolation

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
a = np.array([1,2,3,4,5])
b = np.array([1,4,9,16,25])
```

```python
from scipy.interpolate import interp1d
```

```python
f_linear = interp1d(a, b, kind="linear") #create a linear interpolate function
```

```python
a_new = np.linspace(1, 5, 10)
b_interp = f_linear(a_new)
```

```python
plt.scatter(a, b)
```

```python
plt.scatter(a_new, b_interp)
```

```python
f_cubic = interp1d(a, b, kind="cubic")
```

```python
a_new = np.linspace(1, 5, 10)
b_interp = f_cubic(a_new)
```

```python
plt.scatter(a, b)
```

```python
plt.scatter(a_new, b_interp)
```

### 3. Polynomial Interpolation

```python
import numpy as np 
import matplotlib.pyplot as plt 

m = np.array([1,2,3,4,5])
n = np.array([1,4,9,16,25])
```

```python
plt.scatter(m,n)
```

```python
p = np.polyfit(m,n,4)
```

```python
m_new = np.linspace(1,5,10)
n_interp = np.polyval(p,m_new)
```

```python
plt.scatter(m_new, n_interp)
```

```python
p = np.polyfit(m,n,3)
m_new = np.linspace(1,5,10)
n_interp = np.polyval(p,m_new)
plt.scatter(m_new, n_interp)
```
