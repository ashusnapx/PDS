## Handling Outliers

#### If you can find out these <span style ="background-color:yellow">5 parameters</span>
1. Minimum Value
2. Q1 - 25 Percentile
3. Median
4. Q3 - 75 Percentile
5. Maximum

```python
import numpy as np
```

```python
marks = np.array([9.05, 9.05, 9.50, 7.34, 6.60, 7.45, 1000, 1100])
```

```python
np.percentile(marks, 25) # Q1
```

```python
# i'll create a range between [lower fence <---> higher fence]
q1 = np.percentile(marks, 25) # 25 percentile
```

```python
# their's a library called np.quantile, which will give me all 5 vales
minimum, q1, q2, q3, maximum = np.quantile(marks, [0,0.25,0.50,0.75,1.0])
```

```python
maximum
```

```python
minimum
```

```python
# to find out lower fence, we need to find inter quartile range (IQR)
iqr = q3 - q1
print(iqr)
```

```python
lower_fence = q1 - 1.5 * (iqr)
higher_fence = q3 + 1.5 * (iqr)
```

```python
lower_fence, higher_fence
```

```python
outliers = []
for i in marks:
    if i >= -367.13124999999997 and i <= 631.67875:
        print("This element is not an outlier.")
    else:
        outliers.append(i)
```
