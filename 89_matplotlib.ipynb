```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y)
plt.xlabel("This is x axis")
plt.ylabel("This is y axis")
plt.title("This is done for testing!")
plt.show()
```

```python
colors = np.random.rand(50)
size = 100 * np.random.rand(50)
plt.scatter(x , y , c = colors , s = size, alpha = 0.5)
plt.xlabel("This is x axis")
plt.ylabel("This is y axis")
plt.title("This is done for testing!")
```

```python
x = ['a','b','c','d']
y = np.random.rand(4)
plt.plot(x,y)
plt.xlabel("alphabet")
plt.ylabel("numbers")
plt.title("Testing")
plt.show()
```

```python
x = ['a', 'b', 'c', 'd']
y = np.random.rand(4)
plt.bar(x, y)
plt.xlabel("alphabet")
plt.ylabel("numbers")
plt.title("Testing")
plt.show()
```

```python
x = ['a', 'b', 'c', 'd']
y = np.random.rand(4)
plt.barh(x, y)
plt.ylabel("alphabet")
plt.xlabel("numbers")
plt.title("Testing")
plt.show()
```

```python
x = np.linspace(0,10,200)
y = np.sin(x)
plt.plot(x, y, "--y")
plt.xlabel("numbers")
plt.ylabel("numbers")
plt.title("Testing")
plt.show()
```

```python
x = np.linspace(0, 10, 200)
y = np.sin(x)
plt.plot(x, y, "-.v")
plt.xlabel("numbers")
plt.ylabel("numbers")
plt.title("Testing")
plt.show()
```

```python
plt.figure(figsize=(6, 2))
x = np.linspace(0, 10, 200)
y = np.sin(x)
plt.plot(x, y, "-or")
plt.xlabel("numbers")
plt.ylabel("numbers")
plt.title("Testing")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(6, 2))
x = np.linspace(0, 10, 200)
y = np.sin(x)
plt.plot(x, y, "-+")
plt.xlabel("numbers")
plt.ylabel("numbers")
plt.title("Testing")
plt.grid()
plt.show()
```
