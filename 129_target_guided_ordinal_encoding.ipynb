```python
import pandas as pd
df = pd.DataFrame({
    'model': ['Samsung S21 FE', 'Realme 8 4G', 'Poco X3', 'Poco M2', 'Poco X3', 'Realme 8 4G'],
    'price': [40000, 12000, 14000, 10000, 15000, 13000]
})
```

```python
# step 1 is to calculate the mean price of smartphone model
mean_price = df.groupby('model')['price'].mean().to_dict()
```

```python
# step 2 is to replace each city with its mean price
df['model_encoded'] = df['model'].map(mean_price)
```
